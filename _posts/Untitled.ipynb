```python
import pandas as pd
import numpy as np

dicty = {"nb":(3,5,10),
         "price": (15,25,50)}
df = pd.DataFrame(dicty)
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df.nb,df.price,marker="x",color="blue",s=60)
plt.xlim((0,11))
plt.ylim((0,55))
plt.xlabel("Number of pints")
plt.ylabel("Price in pounds")
plt.show()
```

```python
plt.scatter(df.nb,df.price,marker="x",color="blue",s=60)
plt.plot([0,11],[0,55],'k--',color="red")
plt.xlim((0,11))
plt.ylim((0,55))
plt.xlabel("Number of pints")
plt.ylabel("Price in pounds")
plt.show()
```

```python
plt.scatter(df.nb,df.price,marker="x",color="blue",s=60)
plt.plot([0,11],[0,55],'k--',color="red")
plt.plot([8,8],[0,40],'k--',color="black")
plt.plot([0,8],[40,40],'k--',color="black")
plt.scatter(8,40,marker="x",color="black",s=120)
plt.xlim((0,11))
plt.ylim((0,55))
plt.xlabel("Number of pints")
plt.ylabel("Price in pounds")
plt.show()
```
